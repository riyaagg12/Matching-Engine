```python
import socket

def send_order(asset_id, side, price, quantity):
    client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    port = 56789
    client.connect(('127.0.0.1', port))
    order = f"{asset_id},{side},{price},{quantity}"
    client.send(order.encode())
    
    #used to accumulate the server's responses
    buffer = ""
    while True:
        response = client.recv(4096).decode()
        #Accumulates the response data
        buffer += response
        
        #This checks if there's a newline character in the buffer, indicating the end of a message
        # Process each line in the buffer
        while '\n' in buffer:
            line, buffer = buffer.split('\n', 1)
            line = line.strip()
            if line:
                print(line)
        
        #The loop stops when "Order received and processed" is detected in the response and the buffer is empty, indicating that all messages have been received and processed
        # Stop if the final message was received
        if "Order received and processed" in response and buffer == "":
            break

    client.close()
    
    
orders = [
    ('TCS', 'buy', 10, 50),
    ('TCS', 'buy', 10, 20),
    ('TCS', 'sell', 12, 5),
    ('TCS', 'buy', 8, 10),
    ('TCS', 'buy', 22, 10),
    
    ('TCS', 'sell', 20, 5),
    
    ('TCS', 'buy', 12, 20),
    ('TCS', 'sell', 22, 15),
    ('TCS', 'sell', 12, 5),
    ('TCS', 'buy', 20, 12),
    ('TCS', 'buy', 24, 10),
    ('TCS', 'sell', 10, 10),
    ('TCS', 'sell', 24, 10)
]


for order in orders:
    send_order(order[0], order[1], order[2], order[3])
```
